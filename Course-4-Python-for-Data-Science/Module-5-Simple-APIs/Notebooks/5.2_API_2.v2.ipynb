## Example 1: RandomUser API
- Bellow are Get Methods parameters that we can generate. 
- For more information on the parameters [documentation](https://randomuser.me/documentation).


### RandomUser Get Methods - #Info

- **`get_cell()`**:  
  Returns the cell phone number of the generated user.

- **`get_city()`**:  
  Returns the city name where the user resides.

- **`get_dob()`**:  
  Returns the date of birth of the user.

- **`get_email()`**:  
  Returns the email address of the user.

- **`get_first_name()`**:  
  Returns the first name of the user.

- **`get_full_name()`**:  
  Combines and returns the user's full name (first + last name).

- **`get_gender()`**:  
  Returns the user's gender (e.g., "male" or "female").

- **`get_id()`**:  
  Returns an object or value representing the user's ID information.

- **`get_id_number()`**:  
  Returns the ID number (e.g., a national identification number) of the user.

- **`get_id_type()`**:  
  Returns the type of ID used (e.g., "SSN," "passport").

- **`get_info()`**:  
  Returns general metadata or information about the RandomUser data.

- **`get_last_name()`**:  
  Returns the last name of the user.

- **`get_login_md5()`**:  
  Returns the MD5 hash of the user's login password.

- **`get_login_salt()`**:  
  Returns the salt value used for hashing the user's login password.

- **`get_login_sha1()`**:  
  Returns the SHA-1 hash of the user's login password.

- **`get_login_sha256()`**:  
  Returns the SHA-256 hash of the user's login password.

- **`get_nat()`**:  
  Returns the nationality code of the user (e.g., "US," "GB").

- **`get_password()`**:  
  Returns the raw password generated for the user.

- **`get_phone()`**:  
  Returns the user's landline phone number.

- **`get_picture()`**:  
  Returns a URL to the user's profile picture.

- **`get_postcode()`**:  
  Returns the postal code (or ZIP code) of the user's address.

- **`get_registered()`**:  
  Returns the date and details of the user's registration.

- **`get_state()`**:  
  Returns the state or province where the user lives.

- **`get_street()`**:  
  Returns the street address of the user.

- **`get_username()`**:  
  Returns the generated username for the user.

- **`get_zipcode()`**:  
  Returns the ZIP code of the user (same as `get_postcode()`).

## Example 2: Fruityvice API

Another, more common way to use APIs, is through `requests` library. The next lab, Requests and HTTP, will contain more information about requests.

We will start by importing all required libraries.


```python
import requests  # Imports the library for making HTTP requests
import json  # Imports the library for handling JSON data
```

### Code Breakdown

- **Syntax**:
  - `import requests` imports the `requests` library for making HTTP requests.
  - `import json` imports the `json` library for working with JSON data.

- **What it does**:  
  These lines load the `requests` library to facilitate communication with web APIs and the `json` library to parse and manipulate JSON data, enabling interaction with API responses or other JSON files.


We will obtain the [fruityvice](https://www.fruityvice.com) API data using `requests.get("url")` function. The data is in a json format.


```python
data = requests.get("https://fruityvice.com/api/fruit/all")  # Sends a GET request to fetch data about all fruits
```

### Code Breakdown

- **Syntax**:
  - `import requests` imports the `requests` library for making HTTP requests.
  - `import json` imports the `json` library for working with JSON data.
  - `data = requests.get("https://fruityvice.com/api/fruit/all")` sends a GET request to the specified URL.

- **What it does**:  
  This line fetches data from the Fruityvice API, retrieving information about all available fruits. The response is stored in the variable `data` for further processing.


We will retrieve results using `json.loads()` function.


```python
results = json.loads(data.text)  # Converts the JSON response text into a Python dictionary or list
```

### Code Breakdown

- **Syntax**:
  - `import requests` imports the `requests` library for making HTTP requests.
  - `import json` imports the `json` library for working with JSON data.
  - `data = requests.get("https://fruityvice.com/api/fruit/all")` sends a GET request to the specified URL.
  - `results = json.loads(data.text)` parses the JSON response text into a Python dictionary or list.

- **What it does**:  
  This line converts the raw text response from the Fruityvice API into a Python data structure, such as a list of dictionaries, for easier data manipulation and access.


We will convert our json data into *pandas* data frame. 


```python
pd.DataFrame(results)  # Converts the parsed JSON data into a pandas DataFrame for easy manipulation
```

The result is in a nested json format. The 'nutrition' column contains multiple subcolumns, so the data needs to be 'flattened' or normalized.


```python
df2 = pd.json_normalize(results)  # Flattens nested JSON data into a pandas DataFrame
```

```python
df2.head()
```

df2 = pd.json_normalize(results)` flattens nested JSON data and converts it into a pandas DataFrame.

- **What it does**:  
  This line processes the JSON data to ensure that any nested structures are expanded into separate columns, making the data easier to analyze and work with in a tabular format.

Let's see if we can extract some information from this dataframe. Perhaps, we need to know the family and genus of a cherry.


```python
cherry = df2.loc[df2["name"] == 'Cherry']  # Filters the DataFrame to isolate the row for 'Cherry'
(cherry.iloc[0]['family']), (cherry.iloc[0]['genus'])  # Retrieves the 'family' and 'genus' of the Cherry
```

 - `cherry = df2.loc[df2["name"] == 'Cherry']` filters the DataFrame to get the row where the "name" column equals 'Cherry'.
  - `(cherry.iloc[0]['family']), (cherry.iloc[0]['genus'])` retrieves the "family" and "genus" values of the filtered Cherry row.

- **What it does**:  
  This code isolates the Cherry entry from the DataFrame and extracts its "family" and "genus" details. It ensures precise access to specific attributes of the Cherry record for further use or display.

### Code Breakdown

`data2 = requests.get("https://official-joke-api.appspot.com/jokes/ten")`

2. Retrieve results using `json.loads()` function.


```python
# Write your code here
results2 = json.loads(data2.text)  # Converts the JSON string into a Python list or dictionary
```

### Code Breakdown
